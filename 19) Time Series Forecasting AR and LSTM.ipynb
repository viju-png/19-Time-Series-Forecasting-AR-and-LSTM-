```python
roll=A.rolling(window=20).mean()
```

```python
roll.dropna(inplace=True)
```

```python
roll.shape
```

```python
plt.figure(figsize=(20,10))
plt.plot(A,"red")
plt.plot(roll,"blue")
```

```python
train=roll[:600]
test=roll[214:]
```

```python
train
```

```python
test
```

```python
from statsmodels.tsa.ar_model import AR,AutoReg
```

```python
auto=AutoReg(train,lags=10).fit()
```

```python
pred=auto.predict(start="1969-06-01",end="2019-05-01")
```

```python
pred
```

```python
plt.plot(pred)
```

```python
plt.plot(train)
```

```python
plt.plot(train,"red")
plt.plot(pred,"black")
```

```python
from keras.layers import Dense,LSTM
from keras.models import Sequential
```

```python
ss=Sequential()
```

```python
ss.add(LSTM(10,activation="relu",input_shape=(600,1)))
ss.add(Dense(1))
```

```python
ss.compile(optimizer="adam",loss="mse")
ss.fit(train,test,epochs=10)
```
